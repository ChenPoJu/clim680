```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
import cartopy.feature as feature
from cartopy.util import add_cyclic_point
```

```python
# read data

path = '/scratch/pchen25/clim680_dataset/GeoMeetClimateModels/MidMiocene/atm/'
file_name = 'B.MMIOx2_C5_280_WISOon_aeroZonal_ANN_concat.nc'
ds = xr.open_dataset(path+file_name)

ds
```

```python
# input variables

var = 'TS'
ts = ds[var]

ts
```

```python
# input the first time step of the variable

ts_jan = ds[var][0,:,:]

ts_jan
```

```python
# Make a figure and set its size
fig = plt.figure(figsize=(11,8.5))

# Set the axes using the specified map projection (ex: Robinson)
ax = plt.axes(projection=ccrs.Robinson())

# Add cyclic point to data
data = ts_jan
data, lons = add_cyclic_point(data, coord=ds['lon'])

# Make a filled contour plot
cs = ax.contourf(lons, ds['lat'], data, transform = ccrs.PlateCarree(), cmap='coolwarm', extend='both')

# Draw the borders & coastines for each subplot
ax.add_feature(feature.BORDERS, zorder=2, color='k')
ax.coastlines(zorder=3, color='k')

# Add gridlines
ax.gridlines()

# Add colorbar
#cbar = plt.colorbar(cs, shrink=0.7, orientation='horizontal', label='Surface Temperature (K)')
cax  = fig.add_axes([0.22, 0.2, 0.6, 0.02])
CB   = plt.colorbar(cs, cax = cax, orientation='horizontal')
CB.set_label('Surface Temperature (K)', rotation=0, fontsize=14)

# Add title
ax.set_title('Miocene Simulation (with Piocene Aerosols)', y=1.02, fontsize=18)
;
```
